```python
import pandas as pd
```

```python
excel_file = pd.read_excel('C:/Users/25405/Desktop/大数据知识图谱/神话知识图谱/W0基础数据库【0.6神性人物W0560～W0769】.xlsx','0.6.5常见的典型神性人物W0670～W0769')
```

```python
excel_file
```

```python
excel_file.dropna(subset = ['【实例】'], inplace = True)      # 删除指定列空值的行, subset参数指定列，inplace参数为修改原dataframe
```

```python
excel_file
```

```python
instance = excel_file['【实例】']
instance = instance.astype('str')
c = instance[instance.str.contains('（参见')]      # 找到含有'（参见'的所有行
t1 = list(c)      # 将含有'（参见'的所有元素以列表表示为t1
t2 = list(instance)      # 将instance中的所有元素以列表表示为t1
subtraction = list(set(t2) ^ set(t1))      # 采用列表求差集方法将所有含'（参见'的元素除去
```

```python
result = instance[instance.isin(subtraction)]
result
```

```python
j = 1
for i in result:
    #.to_csv('C:/Users/25405/Desktop/大数据知识图谱/神话知识图谱/实例单句/instance-' + i + '.txt', sep = '\t', index = False)
    with open('C:/Users/25405/Desktop/大数据知识图谱/神话知识图谱/实例单句/instance-' + str(j) + '.txt', 'w', encoding = 'utf-8') as f:   # 代码会自动新建
        f.write(i)
        f.close()
    j += 1
```
